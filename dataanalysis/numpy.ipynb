```python
a = np.arange(10,100,.1)
print(a) #10 se lekar 100 99.9 tak elements with common difference of .1
```

```python
print(a.size)#900 elements in array a
print(a.shape)#sha[e of array-900 rows
print(a.ndim)#dimension count 1-d
```

```python
a2 = a.reshape((30,30))#30 rows and 30 columns
print(a2)
```

```python
print(a2.size)#900 elements in a2
print(a2.shape)#shape of array-30 rows and 30 columns(horizontal rows and verical columns)
print(a2.ndim)#dimension count 2-d
```

```python
a3=a.reshape((30,15,2))#here 30 is each block size,15 is row size and 2 is column size
print(a3)
```

```python
b= np.linspace(1,10,100) #just like arrange
print(b) # 1 se lekar 10 tak 100 random elements
```

```python
a = np.random.randint(1,25,(3,4))# 3 rows and 4 columns with values lying between 1 to 25
b = np.random.randint(1,20,(3,4))# 3 rows and 4 columns with values lying between 1 to 20
print(a)
print(b)
print(a+b)#elementwise addition of square matrix
print(a*b)
print(a-b)
print(a/b)
```

```python
a1=np.random.randint(1,5,(2,2))
b1=np.random.randint(1,10,(2,2))
print(a1)
print(b1)
print(a1*b1)# elementwise multiplicaion pf square matrix
```

```python
print(a.max(),a.mean(),a.min(),a.std(),a.sum()) #std is standard deviation of array "a" containing random elements

print(a.max())
print(a.mean())
print(a.min())
print(a.std()) #std is standard deviation of array "a" containing random elements
print(a.sum()) 
```

```python
print(a.max(axis=0))#columnwise maximum value(each column maximum value)
print(a.max(axis=1))#rowwise maximum value(each row maximum value)
```
