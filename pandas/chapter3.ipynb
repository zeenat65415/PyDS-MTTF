#### data cleaning


```python
cols_to_drops=["Type","Coverage","AREA","REG","DEV"] #columns dropping
df.drop(columns=cols_to_drops,inplace=True)
```

```python
print(df.columns.to_list())
```

```python
#column renaming
df.rename(
    columns={
     "OdName":"Country","AreaName":"Continent",
     "RegName":"Region","DevName":"Status"},
     inplace=True   
)
```

```python
df.head(3)
```

#### add a new column

```python
years=list(range(1980,2014))
print(years)
```

```python
df[years].sum(axis=1)
```

```python
df["Total"]=df[years].sum(axis=1)
```

```python
df.head(3)
```

```python
df[1980].plot(kind="hist",figsize=(10,6))
```

```python
df.set_index("Country",inplace=True)
```
